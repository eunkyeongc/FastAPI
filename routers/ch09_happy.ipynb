# 2) 데이터 병합

```python
city = happy_life['시도'].unique()
city
```

```python
# 리스트로 변환
city = list(city)
city
```

```python
# 데이터프레임으로 생성
happy_merge = pd.DataFrame({
    '시도': city
})
happy_merge.head()
```

```python
happy_life.head()
```

## 시도별 삶의 만족도의 평균

```python
life = happy_life['삶의 만족도'].groupby(by=happy_life['시도']).mean()
life  # 시리즈
```

```python
# '시도'를 중심으로 병합한다
happy_merge = pd.merge(happy_merge, life, on='시도')
happy_merge.head()
```

```python
happy_health.head()
```

## 시도별 건강의 평균

```python
health = happy_health['평균'].groupby(happy_health['시도']).mean()
health
```

```python
happy_merge = pd.merge(happy_merge, health, on='시도')
happy_merge.head()
```

```python
happy_merge = pd.merge(happy_merge, health.rename('건강'), on='시도')
happy_merge.head()
```

```python
happy_merge = happy_merge.drop('평균', axis=1)  # 컬럼(열) 삭제
```

```python
happy_merge.head()
```

```python
happy_safe.head()
```

## 시도별 안전의 평균

```python
safe = happy_safe['평균'].groupby(happy_safe['시도']).mean()
safe
```

```python
happy_merge = pd.merge(happy_merge, safe.rename('안전'), on='시도')
happy_merge.head()
```
